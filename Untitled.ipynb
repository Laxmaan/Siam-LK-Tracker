```python
def get_iou(bb1, bb2):
    """
    Calculate the Intersection over Union (IoU) of two bounding boxes.

    Parameters
    ----------
    bb1 : dict
        Keys: {'x', 'y', 'w', 'h'}
        The (x1, y1) position is at the top left corner,
        the (x2, y2) position is at the bottom right corner
    bb2 : dict
        Keys: {'x', 'y', 'w', 'h'}
        The (x, y) position is at the top left corner,
        the (x2, y2) position is at the bottom right corner

    Returns
    -------
    float
        in [0, 1]
    """


    # determine the coordinates of the intersection rectangle
    x_left = max(bb1['x'], bb2['x'])
    y_top = max(bb1['y'], bb2['y'])
    x_right = min(bb1['x']+bb1['w'], bb2['x']+bb2['w'])
    y_bottom = min(bb1['y']+bb1['h'], bb2['y']+bb2['h'])

    if x_right < x_left or y_bottom < y_top:
        return 0.0

    # The intersection of two axis-aligned bounding boxes is always an
    # axis-aligned bounding box
    intersection_area = (x_right - x_left) * (y_bottom - y_top)

    # compute the area of both AABBs
    bb1_area = (bb1['w']) * (bb1['h'])
    bb2_area = (bb2['w']) * (bb2['h'])

    # compute the intersection over union by taking the intersection
    # area and dividing it by the sum of prediction + ground-truth
    # areas - the interesection area
    iou = intersection_area / float(bb1_area + bb2_area - intersection_area)
    assert iou >= 0.0
    assert iou <= 1.0
    return iou
```

```python
attr_f =open('attr_list.txt','r')
attr_lines = list(map(lambda x: x.strip(),attr_f.readlines()))
attrs={}
for line in attr_lines:
    name,attr = line.split('\t')
    attr=re.split(',|,\s*',attr)
    attrs[name]=[x.strip() for x in attr]
    
attrs
```

```python
results = {}
for i in range(len(gt_dirs)):
    gt_file = os.path.join(gt_dirs[i],"groundtruth_rect.txt")
    fc_pred_file = os.path.join(pred_dirs[1][i],"track_rect.txt")
    lk_file = os.path.join(pred_dirs[0][i],"track_rect.txt")

    seq_name = os.path.basename(pred_dirs[0][i])

    gt = open(gt_file,'r')
    fc = open(fc_pred_file,'r')
    lk = open(lk_file,'r')
    keys = ['x','y','w','h']
    gt_lines = list(map(lambda x: {keys[i]:float(y) for i,y in enumerate(re.split(",| |\t",x.strip()))},gt.readlines()))
    fc_lines = list(map(lambda x: {keys[i]:float(y) for i,y in enumerate(re.split(",| |\t",x.strip()))},fc.readlines()))
    lk_lines = list(map(lambda x: {keys[i]:float(y) for i,y in enumerate(re.split(",| |\t",x.strip()))},lk.readlines()))


    iou_fc = [get_iou(u,v) for u,v in zip(gt_lines,fc_lines)]
    iou_lk = [get_iou(u,v) for u,v in zip(gt_lines,lk_lines)]

    results[seq_name]={"SiamFC":iou_fc, "Siam_LK":iou_lk, "category":attrs[seq_name]}
    lk.close()
    fc.close()
    gt.close()
```

```python
def plot_bar(x,labels,y,y1,results,category='ALL'):
    y_colors = ['#e24a337f']*len(y)
    y1_colors = ['#348abd7f']*len(y1)
    
    if category!='ALL':
        for i,name in enumerate(labels):
            if category not in results[name]['category']:
                y_colors[i]=f'{y_colors[i][:-2]}0f'
                y1_colors[i]=f'{y1_colors[i][:-2]}0f'
    
    plt.bar(x,y,color=y_colors,label='SiamFC')
    plt.bar(x,y1,color=y1_colors,label='SiamLK')
    plt.xticks(ticks=x,labels=labels,rotation=90)
    plt.title(f'Mean IOU for {category} category')
    leg = plt.legend()
    leg.legendHandles[0].set_color('#e24a337f')
    leg.legendHandles[1].set_color('#348abd7f')
    plt.show()
```
